# 2. Ventas por categoría

```python
ventas_categoria = df.groupby("Categoría del Producto")["Precio"].sum().sort_values(ascending=False)
ventas_categoria
```

```python
plt.figure(figsize=(10,5))
ventas_categoria.plot(kind="bar")
plt.title("Ingresos por Categoría de Producto")
plt.xlabel("Categoría")
plt.ylabel("Ingresos Totales")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
ventas_categoria_tienda = (
    df.groupby(["tienda", "Categoría del Producto"])["Precio"]
    .sum()
    .reset_index()
)
ventas_categoria_tienda
```

```python
tabla_pivot = ventas_categoria_tienda.pivot(
    index="Categoría del Producto",
    columns="tienda",
    values="Precio"
)

tabla_pivot.plot(kind="bar", figsize=(12,6))
plt.title("Ingresos por Categoría y Tienda")
plt.xlabel("Categoría")
plt.ylabel("Ingresos")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Tienda")
plt.tight_layout()
plt.show()
```

```python
ventas_categoria_t4 = (
    df[df["tienda"] == "Tienda 4"]
    .groupby("Categoría del Producto")["Precio"]
    .sum()
    .sort_values(ascending=False)
)

ventas_categoria_t4
```

```python
plt.figure(figsize=(8,5))
ventas_categoria_t4.plot(kind="bar")
plt.title("Ingresos por Categoría — Tienda 4")
plt.xlabel("Categoría")
plt.ylabel("Ingresos")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
ventas_producto = df["Producto"].value_counts()
ventas_producto.head(10)
```

```python
top10_productos = ventas_producto.head(10)

plt.figure(figsize=(10,5))
top10_productos.plot(kind="bar")
plt.title("Top 10 Productos Más Vendidos")
plt.xlabel("Producto")
plt.ylabel("Cantidad de Ventas")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
